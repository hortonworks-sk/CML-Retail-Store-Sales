```python
import numpy as np
import pandas as pd
```

```python
cols = ['Store', 'Date', 'CPI']

store = [1,1,2,2]
date = [1,2,1,2]
cpi = np.arange(1,7, dtype=float)

cpi[1:3] = np.nan

store_nos = [1,2]
store_details = ['A', 'B'] 


```

```python
stores = list(zip(store_nos, store_details))
```

```python
features = list(zip(store,date, cpi))
```

```python
features = pd.DataFrame(features, columns=cols)
```

```python
stores = pd.DataFrame(stores, columns=['Store', 'StoreDetails'])
```

```python
features
```

```python
stores
```

```python
store_n_features = features.merge(stores, on=['Store'], how='outer')
store_n_features
```

```python
strs = [1,1,1,1,2,2,2,2]
sale_dates = [1,1,2,2,1,1,2,2] 
sale_depts = [1,2,1,2,1,2,1,2]
sales= [100,150,102,152,200,250,202,250]

sales_matrix = list(zip(srs,sale_depts,sale_dates, sales))
sales_matrix

sales_df = pd.DataFrame(sales_matrix, columns=['Store','Department', 'Date', 'Sales'])
sales_df
```

```python
merged_all = pd.merge(sales_df, store_n_features, how='outer', on=['Store','Date'])
merged_all
```
